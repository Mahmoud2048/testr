```python
import yfinance as yf
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Create a Ticker object for Tesla (TSLA)
tesla_ticker = yf.Ticker("TSLA")

# Use the ticker object to extract data
stock_price = tesla_ticker.info["regularMarketPrice"]

# Extract historical stock data for Tesla
tesla_data = tesla_ticker.history(period="max")
volume = tesla_data["Volume"]

# Print the extracted data
print("Tesla stock price:", stock_price)
print("Tesla trading volume:")
print(volume)

url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"

# Send a GET request to the URL and retrieve the HTML content
response = requests.get(url)
html_data = response.text

# Create a BeautifulSoup object to parse the HTML data
soup = BeautifulSoup(html_data, 'html.parser')

# Find the table with Tesla Quarterly Revenue
table = soup.find('table')

# Extract the table data using pandas read_html function
dfs = pd.read_html(str(table))

# Store the first table in the dataframe as tesla_revenue
tesla_revenue = dfs[0]

tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"", regex=True)
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]

print(tesla_revenue.head(5))
```
